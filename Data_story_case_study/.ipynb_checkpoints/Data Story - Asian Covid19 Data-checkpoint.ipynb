```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import cov
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("data/Latest Covid-19 Data in Asia.csv")
```

```python
df.head(10)
```

```python
df=df.set_index('Country')
```

```python
df.head()
```

```python
df.tail(15)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(index=['Bhutan','Brunei'])
```

```python
sns.pairplot(data=df)
```

```python
print(df['Population'].sort_values(ascending=False))
```

```python
fig, scatter = plt.subplots(figsize = (10,8), dpi = 100)
_=sns.scatterplot(x='Total Cases', y='Total Deaths', data=df, hue='Country',size='Population', legend=False)
plt.yscale("log")
plt.xscale("log")
```

```python
fig, scatter = plt.subplots(figsize = (10,8), dpi = 200)
_=sns.scatterplot(x='Total Cases', y='Total Recovered', data=df, hue='Country',size='Total Deaths', legend=False)
plt.yscale("log")
plt.xscale("log")
```

```python
corr1=pearsonr(df['Total Cases'],df['Total Recovered'])
print(corr1)
```

### We can see that total cases and total recovered are highly correlated, nearly a pearson correlation coefficient of 1. This speaks to a consistent recovery rate for the virus as a whole. We can also see that total deaths increases with total cases, which too, makes sense.

```python
rec_rate= np.mean(df['Total Recovered'])/np.mean(df['Total Cases'])
death_rate = 1- rec_rate
rec_rate, death_rate
```

### We can calculate the average recovery rate, which turns out to be roughly 93.3%, which means that the mortality rate is roughly 6.7%, not accounting for errors in our data.

```python
fig, scatter = plt.subplots(figsize = (10,8), dpi = 100)
_=sns.scatterplot(x='Total Tests', y='Total Cases', data=df, hue='Country',size='Population', legend=False)
plt.yscale("log")
plt.xscale("log")
```

```python
fig, scatter = plt.subplots(figsize = (10,8), dpi = 100)
_=sns.scatterplot(x='Total Tests', y='Total Deaths', data=df, hue='Country',size='Population', legend=False)
plt.yscale("log")
plt.xscale("log")
```

```python
fig = plt.figure(figsize=(18,12))
ax= fig.add_subplot()
ax2=ax.twinx()

width=0.4

df['Total Cases/1M population'].plot(kind='bar', color='red',ax=ax, width=width, position=1, label='Total Cases/1M population')
df['Population'].plot(kind='bar', color='blue',ax=ax2, width=width, position=0, label='Population')


ax.set_ylabel('Cases per 1M Population')
ax2.set_ylabel('Population')
plt.show()
```

### First thing that caught my attention is that China's Total Cases/1M Population is almost non-existent on this graph. Their reported total cases is so small relative to their population, that it hardly registers on this graph. Also some of the highest rate of cases per populus occured in the smalled populated countries, such as Bahrain and the Maldives.


```python
(df['Total Cases/1M population'].sort_values(ascending=True)).plot(kind='barh',figsize=(10,10), 
                                                            title='Total Cases per 1 Million Pop.')
```

```python
(df['Deaths/1M population'].sort_values(ascending=True)).plot(kind='barh',figsize=(10,10), 
                                                            title='Deaths per 1 Million Pop.')
```

### We continue to see the trend of smaller countries being more impacted by the virus relative to their population, but this raises another question regarding China. Even India, the next most populous country, falls in the middle of the pack. Is it a coincidence that China, Taiwan and Hong Kong all fall within the bottom 5 of all major countries in Asia?
