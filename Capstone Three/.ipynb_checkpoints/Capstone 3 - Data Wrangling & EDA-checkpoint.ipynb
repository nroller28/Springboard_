## Feature Engineering

Here we will be engineering some features that may prove useful while modeling.

```python
#creating year, month, week and day of week columns
df['Year']=df.index.year
df['Month']=df.index.month
df['Week'] = df.index.isocalendar().week
df['Weekday']=df.index.weekday
df.tail()
```

```python
#Create a few engineered features
features = pd.DataFrame(index=df.index).sort_index() #creating new dataframe containing new features

features['Returns']=df.Close/df.Open-1 #Percent daily return
features['F02'] = df.Open/df.Close.shift(1)-1
features['Volume Diff Daily']=df.Volume.diff() #daily volume difference from prior day
features['Volume Diff 50']=df.Volume.diff(60) #Volume differential from 60 days prior
features['Volume ROC'] = df.Volume.pct_change() #percent rate of change for volume
features['MA5'] = df.Close.rolling(5).mean().apply(np.log) #Log of 5 day moving average of volume
features['Vol per 200MA'] = df.Volume/df.Volume.rolling(200).mean() #daily volume vs 200 day MA close price
features['Close vs 50MA']= df.Close/df.Close.ewm(span=50).mean()-1#daily closing price vs 50 day exponential MA
```

```python
features.tail()
```

Adding an average daily price feature and it's log tranform and creating new dataframe

```python
ETH = df
ETH['Price']=(df['Close']+df['Open'])/2
ETH['LogPrice'] = ETH['Price'].apply(np.log)
ETH['LogVol']=df.Volume.apply(np.log) #log of daily volume
ETH.head()
```

```python
ETH=ETH.asfreq("D")
df=df.asfreq("D")
```

```python
#aggregate into weekly and monthly resamples
df_weekly_mean = df.resample('W').mean()
df_monthly_mean = df.resample('M').mean()
```

```python
#plot daily, weekly and monthly means for 2020-2021
start, end = '2020-01', '2021-09'
fig, ax = plt.subplots()
ax.plot(df.loc[start:end, 'Close'], marker = '.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(df_weekly_mean.loc[start:end, 'Close'], marker = 'o', linestyle='-', label='Weekly Mean Resample')
ax.plot(df_monthly_mean.loc[start:end, 'Close'], marker = '^', linestyle='--', label='Monthly Mean Resample')
ax.set_yscale('Log')
```

```python
#plot monthly resampled close price against volume
fig, [ax1, ax2] = plt.subplots(2,1, sharex=True,gridspec_kw={'height_ratios': [3,1]})
ax1.plot(df_monthly_mean['Close'], color='black')
ax1.set_ylabel('Monthly Resampled Close Price')
ax1.set_title('Monthly Resampled Close Price & Volume')
ax1.set_yscale('Log')
ax2 = df['Volume'].plot.area(linewidth=0)
ax2.set_ylabel('Volume')
plt.show()
```

```python
#same as above except only from 2019 to present, and weekly resample instead of monthly
fig, [ax1, ax2] = plt.subplots(2,1, sharex=True,gridspec_kw={'height_ratios': [3,1]})
ax1.plot(df_weekly_mean.loc['2019':'2021', 'Close'], color='black')
ax1.set_ylabel('Weekly Resampled Close Price')
ax1.set_yscale('Log')
ax1.set_title('Weekly Resampled Close Price & Volume 2019-Present')
ax2 = df.loc['2019':'2021','Volume'].plot.area(linewidth=0)
ax2.set_ylabel('Volume')
plt.show()
```

```python
#Plotting with rolling averages
close_mean = df['Close'].rolling(30,win_type=None).mean()
ax=close_mean.plot(style=['-','--',':'])
ax.set(yscale='Log', title='Rolling Average Close Price')
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

model = SimpleExpSmoothing(df['Close']).fit(smoothing_level=0.01, optimized=True)

ax=df['Close'].plot(linestyle='-')
ax=model.fittedvalues.plot(linestyle='--', color='blue')
ax.set(yscale='Log', title='Close Price & Exponentially Smoothed Close Price', ylabel='Price $')
```
